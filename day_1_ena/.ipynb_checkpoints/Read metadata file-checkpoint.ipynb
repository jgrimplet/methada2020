# Read Metadata file

This short example shows how it is straightforward to read a metadata CSV file in Python with the **pandas** package to subsequently use it for downstream analysis

```python
import pandas as pd
```

Read the CSV file putting it into a DataFrame

```python
df = pd.read_csv('data-sumission.csv', sep='\t', skiprows=2)
```

```python
df
```

Set *sample_alias* as new index and remove *#template* and *#units* rows

```python
df = df.set_index('sample_alias')
df = df.drop('#template').drop('#units')
```

```python
df
```
